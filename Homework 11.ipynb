```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
from astropy import constants as const
```

```python
def distances(e, theta):
    return (1 - e * np.cos(theta))**-1
```

```python
theta = np.arange(0, 2 * np.pi, .01)



distances1 = distances(0.0, theta)
distances2 = distances(0.5,theta)
distances3 = distances(1.0, theta)
distances4 = distances(1.5, theta)
distances5 = distances(2.0, theta)
distances6 = distances(10.0, theta)


```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='polar')
ax.plot(distances1)
ax.set_rmax(2)
ax.set_rticks([1, 2, 3, 4, 5, 6]) 
ax.set_rlabel_position(-22.5)  
ax.grid(True)

ax.set_title("Orbital (Radial) Distances of Plant to Star Systems", va='bottom')
plt.show()
```

Using a prior derivation completed for homework 10, we find the orbital distance related to the orbital speed by the following relations
$$L = mrv_{t}$$  $$r = \frac{a(1-e^2)}{1-ecos\theta}$$  
$$v = \sqrt{\frac{1+e}{1-e} * \frac{GM}{a}}$$ where a is equal to r. If we solve for the ratio of angular momentum and mass, we can then plug in the above noted relations. We can then solve for the orbital speed. 
$$\frac{L}{m} = [a(1-e^2)*\sqrt{\frac{1+e}{1-e} * \frac{GM}{a}}$$
$$\frac{L^2}{m^2} = [(a(1-e^2))^2*\frac{1+e}{1-e} * \frac{GM}{a}$$
$$\frac{L^2}{GMm^2} = a(1-e^2)$$ Now we can plug in the relation for angular momentum that brings in velocity to which we want to isolate.
$$ \frac{(mvr)^2}{GMm^2} = a(1-e^2)$$  $$\frac{v^2r^2}{GM} = a(1-e^2)$$  $$ v = \sqrt{\frac{GM}{r^2} * a(1-e^2)}$$
We can create a proportionality as follows
$$ v \propto \sqrt{a(1-e^2)}$$ With this proportionality we can find the orbital velocity as a function of the true anomaly and eccentricity assuming that r is equated to a.

```python
dist = distances(e, theta)
def velocities(dist, e, theta):
    c = (dist * ( 1 - (e)**2))**-0.5
    return c
```

```python
plt.plot(velocities(dist,0.0,theta))
```
