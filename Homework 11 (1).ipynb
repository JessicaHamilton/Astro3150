```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
from astropy import constants as const
```

```python
def distances(e, theta):
    return 1.0 / (1.0 - e * np.cos(theta))
```

```python
theta = np.arange(0.01, 2. * np.pi, .01)
#distances = np.arange(0, 2, 0.01)


distances1 = distances(0.0, theta)
distances2 = distances(0.5,theta)
distances3 = distances(1.0, theta)
distances4 = distances(1.5, theta)
distances5 = distances(2.0, theta)
distances6 = distances(10.0, theta)


print (distances3)
```

Using a prior derivation completed for homework 10, we find the orbital distance related to the two different velocities, tangential and radial, by the following relations:

Tangential:
$$L = mrv_{t}$$  $$r = \frac{a(1-e^2)}{1-ecos\theta}$$  
$$v = \sqrt{\frac{1+e}{1-e} * \frac{GM}{a}}$$ where a is equal to r. If we solve for the ratio of angular momentum and mass, we can then plug in the above noted relations. We can then solve for the orbital speed. 
$$(\frac{L}{m})^2 = [a(1-e^2)*\sqrt{\frac{1+e}{1-e} * \frac{GM}{a}}]^2$$
$$\frac{L^2}{m^2} = (a(1-e^2))^2*\frac{1+e}{1-e} * \frac{GM}{a}$$
$$\frac{L^2}{GMm^2} = a(1-e^2)$$ Now we can plug in the relation for angular momentum that brings in velocity to which we want to isolate.
$$ \frac{(mvr)^2}{GMm^2} = a(1-e^2)$$  $$\frac{v^2r^2}{GM} = a(1-e^2)$$  $$ v_{t} = \sqrt{\frac{GM}{r^2} * a(1-e^2)}$$
With this expression we can find the orbital velocity as a function of the true anomaly and eccentricity. Now we can plug in the expression for r to have the equation in terms of \theta and e.
$$ v_{t} = \sqrt{\frac{GM}{(\frac{a(1-e^2)}{1-ecos\theta})^2} * a(1-e^2)}$$
$$ v_{t} = \sqrt{\frac{GM(1-ecos\theta)^2}{a(1-e^2)}}$$
Radial:
$$v_{r} = a(1-e^2)[\frac{d}{dt}(1-ecos\theta)^{-1}]$$
$$v_{r} = a(1-e^2)[-(1-ecos\theta)^-2)(esin\theta*\frac{d\theta}{dt})$$

```python
def velocities(e, theta):
    dist = distances(e,theta)
    c = np.sqrt((dist * ( 1 - (e)**2)))
    return c
e = 0.0,0.5,0.9
theta = np.arange(0.01, 2. * np.pi, .01)
```
