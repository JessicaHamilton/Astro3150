```python
air = np.loadtxt('\\Users\\Jahami2802\\Anaconda\\Astro_Homework/holland_temperature.dat') 
sea = np.loadtxt('\\Users\\Jahami2802\\Anaconda\\Astro_Homework/holland_seawater.dat')
plt.subplot(211)
plt.plot(air, 'b', label='air temp')
plt.plot(sea, 'r', label='sea temp')
plt.legend(loc='best')
plt.ylabel('Temp (C)')
plt.xlim(0, 11)
plt.xticks([])
plt.subplot(212)
plt.plot(air-sea, 'ko')
plt.xticks(np.linspace(0, 11, 12),
           ['jan','feb','mar','apr','may','jun','jul','aug','sep','oct','nov','dec'])
plt.xlim(0, 11)
plt.ylabel('Air - Sea temp (C)');
```
